### 1) Matplotlib Introduction

It is a low level graphics library

```python
import numpy as np
import matplotlib.pyplot as plt #this line imports pyplot module from matplotlib library.pyplot module provides a simple interface for creating and customizing various types of plots, such as line plots, scatter plots, histograms, and more.

#special command to display plots in jupyter notebook
%matplotlib inline
```

```python
#create an array:
x=np.linspace(0,np.pi*4,num=200)
```

```python
y=np.sin(x)
```

```python
plt.plot(x,y)
plt.show()  # to avoid [<matplotlib.lines.Line2D at 0x1873790a490>]
```

```python
#simpler example
x=np.arange(3,9)
x
```

```python
y=np.arange(1,7)
y
```

```python
plt.plot(x,y)
plt.show()
```

```python
##:(for first example)
z=np.cos(x)
plt.plot(x,y,'r',label='sin(x)')
plt.plot(x,z,'g',label='cos(x)')#label is for legend function
plt.xlabel('x')# Label x and y axis
plt.ylabel('f(x)')
plt.xlim(5,10)
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(10,5)) #sizing figure of the graph
plt.subplot(1,2,1) #subplotting a single graph
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(x,z,'g')
plt.tight_layout() #to prevent overlapping of subplots
plt.show()
```

```python
# another way of creating the graph- Object Oriented way

fig=plt.figure()
axse=fig.add_axes([0.1,0.8,0.8,0.8])

axse.plot(x,y,label='sin(x)')
axse.legend()

axse.set_xlabel('x') #set x label
axse.set_ylabel('sinx')

axse.set_title('Sine')

plt.show()
```

### 2) Graphs with Matplotlib

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
col_names=['mpg','cylinders','dis','hp','weight','acc','year','origin','name']
```

```python
df=pd.read_csv(r'D:\auto+mpg\auto-mpg.data', sep='\s+', names=col_names) #cleaning the dataset
```

```python
df
```

#### 1) Histogram

```python
plt.hist(df['mpg'])
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.hist(df['mpg'],bins=20)
plt.xlabel('mpg')
plt.show()
```

#### 2) Scatter plot

```python

plt.figure(figsize=(10,5))
plt.scatter(x=df['weight'],y=df['mpg'],alpha=0.5) #alpha gives transparancy to the dots
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

#### 3) histogram 2D

```python

plt.figure(figsize=(10,5))
plt.hist2d(x=df['weight'],y=df['mpg']) 
plt.xlabel('weight')
plt.ylabel('mpg')
plt.colorbar()
plt.show()
```

#### 4) hexagon graph:


```python

plt.figure(figsize=(10,5))
plt.hexbin(x=df['weight'],y=df['mpg'],gridsize=(35,15),cmap='coolwarm')
plt.xlabel('weight')
plt.ylabel('mpg')
plt.colorbar()
plt.show()
```

### 3)Graphs with Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings

warnings.filterwarnings('ignore')
%matplotlib inline
```

```python
df=pd.read_csv(r'D:\auto+mpg\auto-mpg.data', sep='\s+', names=col_names) #cleaning the dataset
```

```python
df.head()
```

#### 1) Histogram

```python
# produce histogram
plt.figure(figsize=[10,6])
sns.distplot(df['mpg'],kde=False,bins=20) #to avoid the density line,kernel density(kde) is used
plt.show()
```

#### 2) Scatter plot

```python
sns.lmplot(x='weight',y='mpg',data=df,fit_reg=False,aspect=2,height= 4,hue='cylinders')# fit_reg , to remove linear regression line
#aspect for x axis
# height for y axis
# the graph multiplies the points of aspect and height(ex. 2*4=8)
plt.show()
```

```python
# dividing one scatter plot into many columns
sns.lmplot(x='weight',y='mpg',data=df,fit_reg=False,aspect=1,height= 5,hue='cylinders',col='origin')
plt.show()
```

#### 3) Kernel density plot

```python
sns.kdeplot(x=df['weight'],y=df['mpg'])
plt.show()
```

#### 4) Joint plot

```python
sns.jointplot(x=df['weight'],y=df['mpg'],alpha=0.9)
plt.show()
```

#### 5) Pair Grid:


```python
g=sns.PairGrid(df,diag_sharey=False)
g.map_lower(sns.kdeplot)
g.map_upper(plt.scatter)
g.map_diag(sns.kdeplot)
plt.show()
```

#### 6) counter plot

```python
sns.countplot(y='cylinders',data=df)
plt.show()
```

#### 7) Box plot

```python
sns.boxplot(x='origin',y='mpg',data=df,hue='cylinders')
plt.show()
```

### 4) Graphs with Pandas

```python
# same auto dataset
df.tail() # looks at the last few lines
```

#### 1) Histogram

```python
df['weight'].plot(kind='hist',bins=29) #plot function is used to plot graphs in pandas
# kind attribute shows what kind of graph to be plotted
plt.show()
```

```python
# grouping the graph based on number of cylinders
df.groupby('cylinders')['mpg'].plot(kind='hist',alpha=0.4,cmap='coolwarm')
plt.show()
```

#### 2)scatter plot

```python
df.plot(kind='scatter',x='weight',y='mpg',alpha=0.4)
plt.show()
```

#### 3) Boxplot

```python
df.boxplot(column='mpg',by='cylinders',grid=False)

plt.show()
```

#### 4) pivot table

```python
dfp=df.pivot_table(index='cylinders',aggfunc='mean')
```

```python
dfp
```

```python
for col in dfp.columns:
    dfp.plot.line(y=col)
    
plt.show()
```

#### 5) bar plot

```python
df.groupby('cylinders').count()['mpg'].plot(kind='bar')
plt.show()
```
