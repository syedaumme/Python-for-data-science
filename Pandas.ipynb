#### 2) Reading data files

##### Reading csv files

```python
## to create a csv file:
#code:
import csv

# Data to write to the CSV file
data={"name":["john","Sophie","patrick","paul","peter"],
     "gender":["M","F","M","M","M"],
     "country":["USA","england","India","Germany","Russia"],
     "age":[23,18,34,5,31]}


# Name of the CSV file
filename = 'data.csv'

# Writing data to the CSV file
with open(filename, 'w', newline='') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerows(data)

print(f'CSV file {filename} created successfully.')
```

#### .loc - lable based indexing

It is a label based indexer
It allows you to select data based on rows and columns

```python
df.loc[:,['tip']]   # df.loc[rows,[columns]]
```

```python
df.loc[133:243,['tip','sex','size']]
```

```python
#set an index for a row
df2=df.loc[[3,15]]

# setting index
df2=df2.set_index('sex')
df2.head()
df2.loc['female'] #gets rows with only female in it
```

#### .iloc

It is a integer location based indexer

```python
df=df.iloc[0:4,1:5]
```

```python
df.iloc[0:,['tip','day','sex']] #labels wont work here
```
